```python
file_path = "C:/Users/thege/OneDrive/Escritorio/IH/Week 5,6P/data/clean/WebData.csv"
df = pd.read_csv(file_path)
df
```

```python
#Completion Rate: The proportion of users who reach the final ‘confirm’ step.
Percent_cofirm = df['process_step'].value_counts().loc['confirm']/755405
Percent_cofirm
```

```python
df['process_step'].value_counts()
```

```python
#Time Spent on Each Step: The average duration users spend on each step.
df['date_time'] = pd.to_datetime(df['date_time'])

df = df.sort_values(by=['visitor_id', 'visit_id', 'date_time'])

df['time_diff'] = df.groupby(['visitor_id', 'visit_id'])['date_time'].diff()

avg_time_per_step = df.groupby('process_step')['time_diff'].mean()

#Error Rates: If there’s a step where users go back to a previous step, it may indicate confusion or an error. You should consider moving from a later step to an earlier one as an error.
df['next_process_step'] = df.groupby(['visitor_id', 'visit_id'])['process_step'].shift(-1)
df['step_error'] = df['process_step'] > df['next_process_step']

error_rates = df[df['step_error']].groupby('process_step').size() / df.groupby('process_step').size()

print("Average Time Spent on Each Step:")
print(avg_time_per_step)
print("\nError Rates for Each Step:")
print(error_rates)
```
